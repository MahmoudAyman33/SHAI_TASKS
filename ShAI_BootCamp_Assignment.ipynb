# Basic Data Exploration

```python
#Identify the number of rows and columns in the dataset
n_rows = df.shape[0]
print('number of rows =', n_rows)
n_columns = df.shape[1]
print('number of columns =', n_columns)
```

```python
#determine the data types of each column
print("Data types of each  column :",df.dtypes)
```

```python
#check for missing values in each column
print("missing values in each column :\n",df.isnull().sum())
```

# Descriptive Statistics

```python
#Calculate basic statistics mean, median, mode, minimum, and maximum salary
#i'll name each step and put in a variable tp print it later 
statistcs = df['TotalPay'].describe()
```

```python
# determine the range of salaries
s_max = df['TotalPay'].max()
s_min = df['TotalPay'].min()
s_range = s_max - s_min
```

```python
#calculating the standard deviation
sd_salary = df['TotalPay'].std()
```

```python
#printing the results
print('Statistics :\n',statistcs)
print('Salary Range :',s_range)
print('Standerd Deviation :',sd_salary)
```

# Data Cleaning

```python
# Handle missing data by suitable method with explain why you use it.
#first lets see what do i need to with the data 
df.duplicated().sum()
```

```python
#as we see there is no duplicates 
#the missing values :
df.isnull().sum()
```

```python
#we have missing values so lets deal with them
null_values = df.fillna(df.mean() , inplace = True)
```

```python
df.isnull().sum()
```

```python
#the 'Notes' and 'Status' columns are completly null values so ill drop them
df = df.drop(['Notes', 'Status'], axis=1)
```

# Basic Data Visualization

```python
#Create histograms or bar charts to visualize the distribution of salaries,
import matplotlib.pyplot as plt
df['TotalPay'].plot(kind='hist',bins = 25)
plt.title('distribution of salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
#and use pie charts to represent the proportion of employees in different departments
plt.pie(df['JobTitle'].value_counts())
plt.title('Employee Distribution by Department')
```

# Grouped Analysis

```python
#Group the data by one or more columns and calculate summary statistics for each group
grouped_data = df.groupby(['JobTitle', 'Year'])['TotalPay'].agg(['mean', 'median', 'min', 'max'])

print(grouped_data)
```

```python
#and compare the average salaries across different groups.
grouped_data_multi = df.groupby(['JobTitle', 'Year'])['TotalPay'].mean()

print("Average Salary for Each Job Title and Year:")
print(grouped_data_multi)
```

# Simple Correlation Analysis

```python
#: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.
correlation = df['TotalPay'].corr(df['OvertimePay'])
plt.scatter(df['TotalPay'], df['OvertimePay'])
plt.xlabel('Total Pay')
plt.ylabel('OvertimePay')
plt.show()
```
